### No. of Defaulters vs Non-Defaulters

In this section, we analyze the distribution of clients who defaulted on their credit card payments compared to those who did not.

```python
df['default_payment_next_month'].value_counts(normalize=True) * 100
```

### Average Credit Limit for Defaulters vs Non-Defaulters
In this section we analyze whether there is a significant difference in the average credit limit between clients who defaulted on their payments and those who didn't. This will help us comprehend if credit limit plays a role in default behavior - i.e., are clients with higher or lower credit limit more likely to default?

I use a simple groupby function to calculate the average credit limit for each group (0 = non-defaulters, 1 = defaulters).

```python
df.groupby('default_payment_next_month')['limit_bal'].mean()
```

### Default Rate by Education Level
Here I show how the default rates vary across different education levels, highlighting which groups have higher or lower risk.

```python
education_default_rate = df.groupby('education_label')['default_payment_next_month'].mean().multiply(100).round(2)
print(education_default_rate)
```

```python
education_default_rate.plot(kind='bar', color='coral')
plt.title('Default Rate by Education Level')
plt.ylabel('Default Rate (%)')
plt.xlabel('Education Level')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Clients with only a High schoo education show the highest default rate, followed by those with a University background. In contrast, those who attended graduate school have a lower likelihood of defaulting. This suggests a possible link between higher education and improved financial behavior or stability

### Education vs Credit Limit
This section explores how average credit limits differ among clients with varying education levels.

```python
edu_credit = df.groupby('education_label')['limit_bal'].mean().sort_values(ascending = False)
print(edu_credit)
```

```python
edu_credit.plot(kind = 'bar', color = 'teal')
plt.title('Average Credit Limit by Education Level')
plt.ylabel('Average Credit Limit (R)')
plt.xlabel('Education Level')
plt.xticks(rotation = 45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.tight_layout()
plt.show()
```

### Insight:
The bar chart shows that clients with higher education levels such as Graduate school and "other", tend to have higher average credit limits compared to those with lower education levels like High school. This suggests that education may influence creditworthiness or lending decisions.

### Default Rate by Marital Status
The analysis compares the default rates between married and non-married clients. It helps determine whether marital status influences the likelihood of defaulting on payments.

```python
marriage_map = {
    1: 'Married',
    2: 'Single',
    3: 'Other',
    0: 'Unknown'
}
df['marriage'] = df['marriage'].map(marriage_map)
```

##### I changed the numeric codes, of the marriage column, to descriptive labels - for better readability.

```python
default_by_marriage = df.groupby('marriage')['default_payment_next_month'].mean() * 100
default_by_marriage = default_by_marriage.round(2)
```

```python
print(default_by_marriage)
```

```python
#Relatioship between default rate and marital status
default_by_marriage.plot(kind = 'bar', color = 'salmon', figsize = (8, 5))

plt.title('Default Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Default Rate (%)')
plt.xticks(rotation = 0)
plt.grid(axis ='y', linestyle = '--', alpha = 0.6)
plt.tight_layout()
plt.show()
```

#### Insight:
What I observed here is that marital status appears to influence credit default risk, with "Other" (which might include divorced, widowed, or unclassified individuals) and "Married" individuals showing greater risk compared to "Single" individuals.

### Default by Sex/Age
We will compare the default rates between male and female clients to identify any gender-based differences in credit repayment behavior.

```python
default_by_sex = df.groupby('sex')['default_payment_next_month'].mean() * 100
default_by_sex = default_by_sex.round(4)
print(default_by_sex)
```

```python
df.groupby('sex')['default_payment_next_month'].mean().multiply(100).round(2)
```

```python
#default_by_sex
df.groupby('sex')['default_payment_next_month'].mean().multiply(100).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Default Rate by Sex')
plt.ylabel('Default Rate (%)')
plt.xlabel('Sex')
plt.ylim(0, 30)
plt.grid(axis='y', linestyle= '--', alpha = 0.7)
plt.show()
```

```python
#default rate by sex and age
default_by_age_sex = df.groupby(['sex', 'age'])['default_payment_next_month'].mean().multiply(100).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=default_by_age_sex, x='age', y='default_payment_next_month',
             hue = 'sex',
             marker = 'o'
            )
plt.title('Default Rate by Age and Sex')
plt.xlabel('Age')
plt.ylabel('Default Rate (%)')
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

The default rate generally deteriorates with age for both males and females, suggesting younger clients are more likely to default. However, at most age levels, males tend to have a slightly higher default rate than females, indicating potential gender-related differences in credit management.

```python
#correlation netween credit limit and default
df[['limit_bal',
    'default_payment_next_month']].corr()
```

```python
df['credit_bin'] = pd.qcut(df['limit_bal'], q = 5)
```

```python
df.groupby('credit_bin', observed = False)['default_payment_next_month'].mean().multiply(100).round(2)
```

```python
df.groupby('credit_bin', observed = False)['default_payment_next_month'].mean().multiply(100).plot(kind='bar', color = 'teal')
plt.title('Default Rate by Credit Limit Range')
plt.ylabel('Default Rate (%)')
plt.xlabel('Credit Limit Range')
plt.xticks(rotation = 45)
plt.grid(axis = 'y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### Insight:
As credit limits increase, the default rate tends to decrease. Clients with lower credit limits (R50 000 and below) have the highest default rate at 31.79%, while those with the highest credit limits (above R270 000) have the lowest default rate at 13.80%. This suggests that higher creditworthiness may be associated with more responsible repayment behavior.

### Average Repayment Status Over Time
Here we will show the average repayment status across the past 6 months for both defaulters and non-defaulters. A higher repayment status value indicates delayed payments (with positive numbers representing months past due). This helps  observe repayment behavior patterns leading up to default.

```python
repayment_cols = ['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
avg_repayments = df.groupby('default_payment_next_month')[repayment_cols].mean().T
print(avg_repayments)
```

```python
df.columns
```

```python
avg_repayments.plot(kind = 'line', figsize = (10, 6), marker = 'o')
plt.title('Average Repayment Status by Month')
plt.xlabel('Month (pay_0 = Most Recent)')
plt.ylabel('Average Repayment Status')
plt.legend(['No Default', 'Defaulted'])
plt.grid(True)
plt.show()
```

Defaulters consistenly show higher repayment status values compared to non-defaulters, especially in the most recent months. This might suggest that repayment difficultues begin months before the actual default and could serve as early warning indicators.





### Education Level: Credit Limit vs Default Rate
This section compares average credit limits and default rates across education levels to explore how education correlates with creditworthiness and repayment risk.

```python
edu_credit_default = pd.DataFrame({
    'Average Credit Limit': edu_credit.round(2),
    'Default Rate (%)': education_default_rate.round(2)
}).reset_index()

print(edu_credit_default)
```

```python
edu_credit = df.groupby('education_label')['limit_bal'].mean()
education_default = df.groupby('education_label')['default_payment_next_month'].mean().multiply(100).round(2)

edu_credit_default = pd.DataFrame({
    'Average Credit Limit': edu_credit,
    'Default Rate (%)': education_default
})
```

```python
 fig, ax1 = plt.subplots(figsize=(10,6))

#First bar: Avg Credit Limit
ax1.bar(edu_credit_default.index, edu_credit_default['Average Credit Limit'], color = 'skyblue', label = 'Avg Credit Limit')
ax1.set_ylabel('Average Credit Limit (R)', color = 'skyblue', fontsize = 12)
ax1.tick_params(axis = 'y', labelcolor = 'skyblue')

#Second y-axis
ax2 = ax1.twinx()
ax2.plot(edu_credit_default.index, edu_credit_default['Default Rate (%)'], color = 'crimson', marker = 'o', label = 'Default Rate (%)')
ax2.set_ylabel('Default Rate (%)', color = 'crimson', fontsize = 12)
ax2.tick_params(axis = 'y', labelcolor = 'crimson')

#Title and formatting
plt.title('Education Level: Credit Limit vs Default Rate', fontsize = 14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
fig.tight_layout()
plt.savefig("Education Level: Credit Limit vs Default Rate.png", dpi = 300, bbox_inches='tight')
plt.show()
```

### Insight
From the bar chart it can be observed that High Schools graduates have the lowest credit access of approx. R126 550 but have the highest default rate of 25.16%. Implying that they struggle with repayment capacity. Now if we look at "Other" education group, we see that they have the highest credit limit, approx. R220 894 and have the lowest default rate of 5.69%. These are the ideal clients for banks/financial institutions.

## Overall Summary of Insights

This analysis explored various factors influencing credit default among clients. Below, I have included some of the findings that I observed from the dataset:

Defaulter Distribution: A significant portion of the dataset consists of non-defaulters, however the default rate remains notable and non-negligible.

Credit Limit & Default Rate: There is a clear inverse or indirect relationship between the two — as credit limits increase, default rates tend to decrease. Clients with lower credit limits are more likely to default.

Education Level: Clients with higher education levels (Graduate School) tend to have higher credit limits and lower default rates, immplying a potential link between education, financial literacy, and creditworthiness.

Marital Status: Individuals classified under "Other" and "Married" statuses show higher default rates, whereas Single clients have the lowest. This may reflect varying financial responsibilities or stability.

Gender and Age: Default patterns vary slightly by sex and age. Younger individuals show slightly higher default tendencies, and males tend to default more than females, though differences aren't dramatic.

Repayment Trends: Across past months, average repayment status worsens the further back in time we look, especially for defaulters — indicating longer-term payment struggles.

### Conclusion:
The default risk appears to be influenced by a combination of demographic (education, marital status, sex, age) and financial (credit limit, past repayment behaviour) factors. These insights can help financial institutions in risk profiling, targeted credit offerings, and proactive credit management strategies.
